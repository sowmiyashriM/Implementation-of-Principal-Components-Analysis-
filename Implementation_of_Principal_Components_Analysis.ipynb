```python
X=df_dummy.drop('price',axis=1)
Y=df_dummy.price
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

prcoan = PCA(n_components = 2)

X_train = prcoan.fit_transform(X_train)
X_test = prcoan.transform(X_test)

explained_variance = prcoan.explained_variance_ratio_
```

```python
total_variance = prcoan.explained_variance_ratio_.sum() * 100
```

```python
total_variance
```

```python
labels = {str(i): f"PC {i+1}" for i in range(2)}
```

```python
import plotly.express as px
fig = px.scatter_matrix(
    X_train,

    dimensions=range(2),
    labels=labels,
    title=f'Tot Explained Variance: {total_variance:.2f}%',
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
